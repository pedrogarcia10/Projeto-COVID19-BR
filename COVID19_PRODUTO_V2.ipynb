## Previsão para os próximos x dias por estado

```python
#Importando módulos
import pandas as pd
import joblib
import datetime

#Carregando o modelo
model = joblib.load("model_RID.pkl")
model

#Carregando dados
records = [(datetime.date(2020, 4, 18), 'RO', 110), (datetime.date(2020, 4, 18), 'AC', 142), (datetime.date(2020, 4, 18), 'AM', 1897),
           (datetime.date(2020, 4, 18), 'RR', 201), (datetime.date(2020, 4, 18), 'PA', 640), (datetime.date(2020, 4, 18), 'AP', 393),
           (datetime.date(2020, 4, 18), 'TO', 33), (datetime.date(2020, 4, 18), 'MA', 1040), (datetime.date(2020, 4, 18), 'PI', 123),
           (datetime.date(2020, 4, 18), 'CE', 3034), (datetime.date(2020, 4, 18), 'RN', 516), (datetime.date(2020, 4, 18), 'PB', 205),
           (datetime.date(2020, 4, 18), 'PE', 2193), (datetime.date(2020, 4, 18), 'AL', 132), (datetime.date(2020, 4, 18), 'SE', 71),
           (datetime.date(2020, 4, 18), 'BA', 1193), (datetime.date(2020, 4, 18), 'MG', 1077), (datetime.date(2020, 4, 18), 'ES', 952),
           (datetime.date(2020, 4, 18), 'RJ', 4543), (datetime.date(2020, 4, 18), 'SP', 13894), (datetime.date(2020, 4, 18), 'PR', 945),
           (datetime.date(2020, 4, 18), 'SC', 962), (datetime.date(2020, 4, 18), 'RS', 831), (datetime.date(2020, 4, 18), 'MS', 161),
           (datetime.date(2020, 4, 18), 'MT', 171), (datetime.date(2020, 4, 18), 'GO', 378), (datetime.date(2020, 4, 18), 'DF', 762)]

#Copiando dados
forecast = records.copy()

#Definindo o número de dias para previsão
dias = 7

#Criando range de datas para forecast
datas = pd.date_range(records[0][0],periods=dias+1)

#Lista de tuplas com a previsão dos dias seguintes
output = []

for dia in range(0,dias):
    
    new_forecast = []
    
    for uf in range(0,len(forecast)):
        
        #Features do modelo em ordem de implementação no predict
        #'Cases_Last_Day', 'UF_CAT_TO', 'UF_CAT_SE', 'UF_CAT_AL', 'UF_CAT_MS', 'UF_CAT_GO', 'UF_CAT_AC'
        cld = forecast[uf][2]
        to = 1 if forecast[uf][1] == 'TO' else 0
        se = 1 if forecast[uf][1] == 'SE' else 0
        al = 1 if forecast[uf][1] == 'AL' else 0
        ms = 1 if forecast[uf][1] == 'MS' else 0        
        go = 1 if forecast[uf][1] == 'GO' else 0
        ac = 1 if forecast[uf][1] == 'AC' else 0

        pred = model.predict([[cld,to,se,al,ms,go,ac]])[0]
                
        att = (datas[dia+1].date(),forecast[uf][1],pred.round().astype(int))
        output.append(att)
        new_forecast.append(att)
        
    forecast = new_forecast
    
output
```
